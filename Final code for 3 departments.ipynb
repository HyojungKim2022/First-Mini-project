```python
# MAKE A TABLE FOR SHILLA DUTY FREE

import pymysql

db = pymysql.connect(host='localhost', port=3306, user='root', passwd='1234', db='project1', charset='utf8')
ecommerce = db.cursor()

sql = """
    create table shilla  (
        ranks int,
        dept_name varchar(20) not null,
        brand_name varchar(200) not null,
        item_name varchar(200) not null,
        dollar_price varchar(20),
        won_price varchar(20),
        before_dist varchar(20),
        dist_percentage varchar(20),
        primary key(ranks,dept_name)
    );
"""
ecommerce.execute(sql)
```

```python
# SHILLA DUTY FREE TABLE CODE

options = webdriver.ChromeOptions()
options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36')

url = 'https://www.shilladfs.com/estore/kr/ko/c/2'
driver = webdriver.Chrome(options=options)
driver.get(url)

elem = driver.find_element_by_xpath('//*[@id="facetListCount"]')
elem.click()
elems = driver.find_element_by_xpath('//*[@id="facetListCount"]/option[4]')
elems.click()
```

```python
# SHILLA DUTY FREE TABLE CODE

dept_name = driver.find_element_by_xpath('//*[@id="footer"]/div[2]/div/div[1]/strong').text
totals = driver.find_elements_by_css_selector('div.pro_detail')
ranks = 1
for total in totals:
    if total.find_element_by_css_selector('strong.info_brand').text == '':
        continue
    
    brand_name = total.find_element_by_css_selector('strong.info_brand').text
    item_name = total.find_element_by_css_selector('p.info_name').text
    dollar_price = total.find_element_by_css_selector('p.p_sum>strong').text
    won_price = total.find_element_by_css_selector('p.p_sum>i').text
    try:
        before_dist = total.find_element_by_css_selector('i.strike').text
        dist_percentage = total.find_element_by_css_selector('p.p_rate>strong').text
    except: 
        before_dist = None
        dist_percentage = None
            
    sql = """INSERT INTO shilla VALUES(
    {0}, "{1}", "{2}", "{3}", "{4}", "{5}", "{6}", "{7}");""".format(ranks, dept_name, brand_name, item_name, dollar_price, won_price, before_dist, dist_percentage)
    ranks += 1
    print (sql)
    ecommerce.execute(sql)
```

```python
# SHINSEGAE DUTY FREE CODE

options = webdriver.ChromeOptions()
options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36')

for page in range(0,160,160):
    url = 'https://www.ssgdfs.com/kr/dispctg/initDispCtg?collection=goods&startCount={}&listCount=160&sort=SALE_CNT%2FDESC&sortNm=%ED%8C%90%EB%A7%A4%EB%9F%89%EC%88%9C&viewType=thumb&filterBrand=&filterType=&filterEtc=&filterPrice=&filterDetail=&filterColor=&filterKeyword=&langCd=KR&disp_ctg_no=2105008627&myFilterOn='.format(page)
    driver = webdriver.Chrome(options=options)
    driver.get(url)

    dept_name = driver.find_element_by_xpath('/html/body/footer/div/p[1]').text
    totals = driver.find_elements_by_css_selector('li.prodCont')
    ranks = 1
    for total in totals:
        if total.find_element_by_css_selector('span.brandName').text == '':
            continue
            
        kor_title = total.find_element_by_css_selector('span.brandName').text
        eng_title = total.find_element_by_css_selector('span.eng').text
        brand_name = kor_title[:len(kor_title) - len(eng_title)]
        item_name = total.find_element_by_css_selector('em.prodName').text
        dollar_price = total.find_element_by_css_selector('strong.saleDollar').text
        won_price = total.find_element_by_css_selector('em.saleWon').text
        try:
            before_dist = total.find_element_by_css_selector('span.originPrice').text
            dist_percentage = total.find_element_by_css_selector('strong.saleNum').text
        except: 
            before_dist = None
            dist_percentage = None

        sql = """INSERT INTO shinsegae VALUES(
        {0}, "{1}", "{2}", "{3}", "{4}", "{5}", "{6}", "{7}");""".format(ranks, dept_name, brand_name, item_name, dollar_price, won_price, before_dist, dist_percentage)
        ranks += 1
        print (sql)
        ecommerce.execute(sql)
```
